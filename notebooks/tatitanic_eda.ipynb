```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('titanic')

# Basic info
print(df.info())
print(df.isnull().sum())
print(df.describe())
```

```python
# Survival Rate by Sex
df.groupby('sex')['survived'].mean()
```

```python
# Survival Rate by Class
df.groupby('sex')['survived'].mean()
```

```python
# Survival Rate by Age Buckets
df['age_group'] = pd.cut(df['age'], bins=[0, 18, 35, 50, 100], labels=['Child', 'Young Adult', 'Adult', 'Senior'])
df.groupby('age_group')['survived'].mean()
```

```python
# Visualization
# Bar Plot – Survival by Sex
sns.barplot(x='sex', y='survived', data=df)
plt.title('Survival Rate by Sex')
plt.show()
```

```python
# Survival Rate by Class
sns.barplot(x='class', y='survived', data=df)
plt.title("Survival Rate by Class")
plt.show()
```

```python
# Boxplot – Age vs Survival
sns.boxplot(x='survived', y='age', data=df)
plt.title('Age Distribution by Survival')
plt.show()
```

```python
# Violin Plot – Age vs Class & Survival
sns.violinplot(x='class', y='age', hue='survived', data=df, split=True)
plt.title('Age Distribution by Class and Survival')
plt.show()
```
